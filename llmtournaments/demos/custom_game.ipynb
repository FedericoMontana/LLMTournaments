```python
import re

var = 'hi: "lala lolo"'
print(var)
match = re.match(r"(.+):\s*(.+)", var)
if match:
    recipient_name, message = match.group(1), match.group(2)
    print(recipient_name, "--", message)
```

```python
import os

from llmtournaments.custom_game import BettingGame, LLMPlayer
from llmtournaments.llm.generators.abacus_api import AbacusLLMInteraction

# from llmtournaments.llm.generators.dummy_llm_02 import DummyLLMInteraction02

num_players = 3
threshold = 500
initial_balance = 1000
rounds = 5

abacus_key = os.getenv("ABACUSAI_API_KEY")

player1 = LLMPlayer(
    llm=AbacusLLMInteraction(abacus_key, llm_name="OPENAI_GPT4O_MINI", temperature=1),
    name="ChatGPT",
    balance=initial_balance,
)

player2 = LLMPlayer(
    llm=AbacusLLMInteraction(abacus_key, llm_name="CLAUDE_V3_5_SONNET", temperature=1),
    name="Claude",
    balance=initial_balance,
)

player3 = LLMPlayer(
    llm=AbacusLLMInteraction(abacus_key, llm_name="LLAMA3_1_70B", temperature=1),
    name="Llama",
    balance=initial_balance,
)

player4 = LLMPlayer(
    llm=AbacusLLMInteraction(abacus_key, llm_name="GEMINI_1_5_PRO", temperature=1),
    name="Gemini",
    balance=initial_balance,
)

# dummy_llm1 = DummyLLMInteraction02(
#     player_number=1, num_players=num_players, threshold=threshold
# )
# dummy_llm2 = DummyLLMInteraction02(
#     player_number=2, num_players=num_players, threshold=threshold
# )
# dummy_llm3 = DummyLLMInteraction02(
#     player_number=3, num_players=num_players, threshold=threshold
# )

# player1 = LLMPlayer(llm=dummy_llm1, name="Player1", initial_balance=initial_balance)
# player2 = LLMPlayer(llm=dummy_llm2, name="Player2", initial_balance=initial_balance)
# player3 = LLMPlayer(llm=dummy_llm3, name="Player3", initial_balance=initial_balance)

game = BettingGame(
    players=[player1, player2, player3, player4],
    total_rounds=rounds,
    threshold=threshold,
    initial_balance=initial_balance,
    max_communication_cycles=3,
)

final_balances = game.run_game()
```
